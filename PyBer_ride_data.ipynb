```python
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
from matplotlib.ticker import MultipleLocator
import numpy as np
import pandas as pd
```

```python
#REad the file
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df

```

```python

# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Bar Chart usin dataframes
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind ='bar')

stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
print(stdev)

plt.show()
```

```python
# Bar Chart usin dataframes

stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", yerr =stdev, capsize =3, color='lightblue')

y_axis = np.arange(0,max(pyber_ride_df["Avg. Fare ($USD)"])+10,step=5.0)
tick_locations = [value for value in y_axis]
print(tick_locations)
print(stdev)
#plt.errorbar(pyber_ride_df["Month"],pyber_ride_df["Avg. Fare ($USD)"], yerr=stdev)
#plt.yticks(np.arange(0, max(pyber_ride_df["Avg. Fare ($USD)"])+10 ,step = 5.0))

plt.yticks(tick_locations)
plt.show()
```
