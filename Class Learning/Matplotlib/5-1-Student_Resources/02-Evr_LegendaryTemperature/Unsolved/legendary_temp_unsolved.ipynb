```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create an array for the x axis for the numerical value of each month 
x_axis = np.arange(0,5,0.1)
x_axis
```

```python
# Create exponentilal serire

e_x = [np.exp(x) for x in x_axis]
e_x
```

```python
#plot
plt.plot(x_axis,e_x)
plt.show()
```

```python
# Avearge high temperature for each month
highs = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Avearge low temperature for each month
lows = [10, 22, 30, 40, 50, 60, 65, 63, 55, 42, 34, 25]
```

```python
# Plot both of these lines so that they will appear on our final chart.
months = np.arange(1,13,1)
months
# Create labels for the X and Y axis

plt.xlabel( 'Months')
plt.ylabel('Temperature (F)')

# Save and display the chart
plt.plot(months,lows, label = "Lows")
plt.plot(months,highs, label = "High")


plt.legend()

plt.show()
plt.savefig("../Images/first.png")
```
