```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
```

```python
# Create a bar chart based upon the above data
plt.bar(x_axis,bars_in_cities, color='b',alpha =0.5, align="center")
 
```

```python
# Create the ticks for our bar chart's x axis
ticks_cities = np.arange(len(cities))
plt.xticks(ticks_cities,cities) 
```

```python
# Set the limits of the x axis
plt.xlim(-0.75, len(x_axis) + .25 )
```

```python
# Set the limits of the y axis
plt.ylim(0, max(bars_in_cities) + 1 )
```

```python
# Give the chart a title, x label, and y label
plt.bar(x_axis,bars_in_cities, color='r',alpha =0.5, align="center")
ticks_cities = np.arange(len(cities))
plt.xticks(ticks_cities,cities) 
plt.title("# of Bars in Cities", color ='w')
plt.xlabel("Cities")
plt.ylabel("# of Bars")
```

```python
# Save an image of the chart and print it to the screen
plt.bar(x_axis,bars_in_cities, color='r',alpha =0.5, align="center")
ticks_cities = np.arange(len(cities))
plt.xticks(ticks_cities,cities, color ='y') 
plt.title("# of Bars in Cities", color ='w')
plt.xlabel("Cities", color ='w')
plt.ylabel("# of Bars", color ='w')
plt.savefig("../Images/barsincity.png")
```
